```python
#receive lat and lng (x, y)
#identify which school district
import requests
import json
import pandas as pd
import csv
import os
import time

URL = 'https://api.schooldigger.com/v1.1/schools?'
```

```python
# Import CSV, put into DataFrame
masterData_csv = pd.read_csv("./Data/latlon_schools_NA.csv")
master_df = pd.DataFrame(masterData_csv)

# Loop through Requests (Partial Params):
alat = []
alon = []
s_name = []
s_district = []
s_county = []

counter = 1

for row in master_df.itertuples():
    params = {
        'st' : 'tx',
        'level' : 'Elementary',
        "nearLatitude": row[3],
        "nearLongitude": row[4],
        'appID' : 'dcf91ee4',
        'appKey' : '9fdecf6c0a6cc7f215a510113300ab20',
        'sortby' : 'distance'
    }
    
    time.sleep(2)
    school_list = requests.get(URL, params=params).json()
    
    alat.append(row[4])
    alon.append(row[5])
    
    try:
        s_name.append(school_list['schoolList'][0]['schoolName'].upper())
    except (KeyError, IndexError):
        s_name.append("NA")
    try:
        county = school_list['schoolList'][0]['county']['countyName'] + ' County'
        county = county.upper()
        s_county.append(county)
    except (KeyError, IndexError, AttributeError, TypeError):
        s_county.append("NA")
        
    print(f"Processed Row # {counter}.")
    counter += 1
    
print(f"Finished Requests.")
```

```python
# Create a DF from Request, 
s_latlon_df = pd.DataFrame({
    "school name": s_name,
    "school county": s_county,
    "alat": alat,
    "alon": alon
})
s_latlon_df.head()

# Write to CSV
s_latlon_df.to_csv("./Data/latlon_schools_02.csv", index=False, header=True)
```
